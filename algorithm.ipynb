```python
import pandas as pd
import folium as f
```

```python
file = 'data\최강지존_박성민.csv'
file = file.replace('\\','//')
cctv = pd.read_csv(file,encoding='cp949')
print(cctv.head(5))
```

```python
x = []
y = []

for i in range(len(cctv['x'])):
    x.append(cctv['x'][i])
    y.append(cctv['y'][i])

print('x갯수: ',len(x))
print('y갯수: ',len(y))
```

```python
# 위도
latitude = 37.394946
# 경도
longitude = 127.111104

m = f.Map(location=[latitude,longitude], zoom_start=17, width=750, height= 500)
m
```

```python
for i in range(len(x)):
    f.Marker([x[i],y[i]]).add_to(m)
m
```
